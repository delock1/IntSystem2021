# Часть 2. Линейная регрессия

```python
%pylab inline

import seaborn as sns
sns.set(style="whitegrid")

from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
```

```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

Постройте график зависимости $R^2$ для обучающей и тестовой (на кросс-валидации) выборок в зависимости от значения гиперпараметра. Для Lasso также постройте график зависимости количества ненулевых весов.

```python
#  Your code here
from sklearn.model_selection import cross_validate
estimators = [Lasso, Ridge]
f, axes = plt.subplots(1, 2, figsize=(13, 5))
for i, estimator in enumerate(estimators):
    test_cv_scores = []
    train_cv_scores = []
    test_scores = []
    for a in alpha:
        regr = estimator(alpha=a)
        cv = cross_validate(regr, x_train, y_train, cv=10, scoring='r2', return_train_score=True)
        test_cv_scores.append(cv['test_score'].mean())
        train_cv_scores.append(cv['train_score'].mean())
        test_scores.append(regr.fit(x_train, y_train).score(x_test, y_test))
    ax = axes[i] 
    ax.set_title(estimator.__name__)
    ax.plot(np.log(alpha), test_cv_scores, label='Score on CV-test')
    ax.plot(np.log(alpha), train_cv_scores, label='Score on CV-train')
    ax.plot(np.log(alpha), test_scores, label='Score on test')
    ax.set_xlabel("log scaled alpha")
    ax.set_ylabel("R2 score")
    ax.grid()
    ax.legend()
```

__Задание 5.__ Оптимальная архитектура, анализ.

__(2 балла)__

При помощи `Pipeline` и `GridSearchSCV` выберите оптимальную архитектуру, комбинируя различные методы масштабирования, степень полинома в `PolynomialFeatures`, а также регуляризаторы. Для `Lasso` поставьте максимальное количество итераций больше значения по умолчанию, чтобы оптимизация сошлась.

```python
#  Your code here
# ...
degrees = list(range(1, 5))
def get_best_models(x_train, y_train):
    pipe = Pipeline(
            [('polynom', PolynomialFeatures()),
             ('scaler', StandardScaler()),
             ('regressor', Lasso())])
    search_params = [
        {
            'polynom': [PolynomialFeatures()],
            'polynom__degree' : degrees,
            'scaler': [StandardScaler(), MinMaxScaler()],
            'regressor':[Lasso(), Ridge()],
            'regressor__max_iter' : [2500],
            'regressor__alpha': alpha
        }
    ]
    gs_cv = GridSearchCV(pipe, param_grid=search_params, refit=True, cv=5, return_train_score=True)
    gs_cv.fit(x_train, y_train)
    return gs_cv
gs_cv = get_best_models(x_train, y_train)
```
